```python
# ==========================================
# E-Commerce Customer Order Analysis
# ==========================================

# Customer Names
customer_names = [
    'Jennie',
    'Sana',
    'Gen',
    'Lisa',
    'Jung',
    'Kook'
]

# Order Details
# Format: (Customer Name, Product, Price, Category)

order_details = [
    ('Jennie', 'Laptop', 1200, 'Electronics'),
    ('Jennie', 'Mouse', 25, 'Electronics'),
    ('Sana', 'T-shirt', 20, 'Clothing'),
    ('Gen', 'Jeans', 25, 'Clothing'),
    ('Lisa', 'Headphones', 50, 'Electronics'),
    ('Sana', 'Smartphone', 800, 'Electronics'),
    ('Jung', 'Coffee Maker', 80, 'Home Essentials'),
    ('Jung', 'Book', 15, 'Books'),
    ('Kook', 'Pen', 5, 'Stationery'),
    ('Gen', 'Monitor', 150, 'Electronics'),
    ('Lisa', 'T-Shirt', 20, 'Clothing')
]

# ==========================================
# Customer Orders Dictionary
# ==========================================

customer_orders_dict = {}

for name, product, price, category in order_details:

    if name not in customer_orders_dict:
        customer_orders_dict[name] = []

    customer_orders_dict[name].append(product)

# ==========================================
# Product Classification
# ==========================================

# Product -> Category Mapping

product_category_map = {
    product: category
    for name, product, price, category in order_details
}

# Unique Categories

unique_categories = {
    category
    for name, product, price, category in order_details
}

print("----- Product Categories Available -----")
print(unique_categories)
print()

# ==========================================
# Customer Spending Analysis
# ==========================================

customer_spending = {}

for name, product, price, category in order_details:

    if name in customer_spending:
        customer_spending[name] += price
    else:
        customer_spending[name] = price

# ==========================================
# Customer Classification
# ==========================================

customer_classes = {}

for name, total in customer_spending.items():

    if total > 100:
        customer_classes[name] = "High Value"

    elif 50 <= total <= 100:
        customer_classes[name] = "Moderate"

    else:
        customer_classes[name] = "Low Value"

# ==========================================
# Revenue Per Category
# ==========================================

category_revenue = {}

for name, product, price, category in order_details:
    category_revenue[category] = (
        category_revenue.get(category, 0) + price
    )

# ==========================================
# Unique Products
# ==========================================

unique_products = {
    product
    for name, product, price, category in order_details
}

# ==========================================
# Electronics Buyers
# ==========================================

electronics_buyers = list(
    set(
        name
        for name, product, price, category
        in order_details
        if category == 'Electronics'
    )
)

# ==========================================
# Top 3 Customers
# ==========================================

top_spenders = sorted(
    customer_spending.items(),
    key=lambda x: x[1],
    reverse=True
)[:3]

# ==========================================
# Customer Report
# ==========================================

print("----- Customer Spending & Classification -----")
print(f"{'Customer':<10} | {'Total Spend':<12} | {'Class'}")
print("-" * 45)

for name, total in customer_spending.items():
    print(
        f"{name:<10} | ${total:<11} | {customer_classes[name]}"
    )

print()

# ==========================================
# Category Revenue
# ==========================================

print("----- Category Revenue -----")

for category, revenue in category_revenue.items():
    print(f"{category}: ${revenue}")

print()

# ==========================================
# Top Customers
# ==========================================

print("----- Top 3 High Value Customers -----")

for i, (name, spend) in enumerate(top_spenders, start=1):
    print(f"{i}. {name} (${spend})")

print()

# ==========================================
# Advanced Set Operations
# ==========================================

# Customers buying from multiple categories

multi_category_buyers = []

for name in customer_names:

    customer_categories = {
        category
        for n, product, price, category in order_details
        if n == name
    }

    if len(customer_categories) > 1:
        multi_category_buyers.append(name)

# Customers buying Electronics

electronics_set = {
    name
    for name, product, price, category
    in order_details
    if category == 'Electronics'
}

# Customers buying Clothing

clothing_set = {
    name
    for name, product, price, category
    in order_details
    if category == 'Clothing'
}

# Intersection

both_category_buyers = (
    electronics_set.intersection(clothing_set)
)

print("----- Advanced Insights -----")

print(
    f"Customers buying from multiple categories: {multi_category_buyers}"
)

print(
    f"Customers who bought BOTH Electronics and Clothing: {both_category_buyers}"
)

print()

# ==========================================
# Additional Insights
# ==========================================

print("----- Electronics Buyers -----")
print(electronics_buyers)

print()

print("----- Unique Products -----")
print(unique_products)


## Customer Spending Analysis
import matplotlib.pyplot as plt

customers = list(customer_spending.keys())
spending = list(customer_spending.values())

plt.figure(figsize=(8,5))
plt.bar(customers, spending, color='skyblue')

plt.title("Customer Spending Analysis")
plt.xlabel("Customer")
plt.ylabel("Total Spend ($)")

plt.tight_layout()

plt.savefig("customer_spending.png")
plt.show()


##Revenue by category
categories = list(category_revenue.keys())
revenues = list(category_revenue.values())

plt.figure(figsize=(8,5))
plt.bar(categories, revenues, color='green')

plt.title("Revenue by Product Category")
plt.xlabel("Category")
plt.ylabel("Revenue ($)")

plt.xticks(rotation=45)

plt.tight_layout()
plt.savefig("category_revenue.png")
plt.show()

## Customer Segmentation
segment_count = {}

for category in customer_classes.values():
    segment_count[category] = segment_count.get(category, 0) + 1

plt.figure(figsize=(6,6))

plt.pie(
    segment_count.values(),
    labels=segment_count.keys(),
    autopct='%1.1f%%'
)

plt.title("Customer Segmentation")
plt.savefig("customer_spending.png")
plt.show()
```
